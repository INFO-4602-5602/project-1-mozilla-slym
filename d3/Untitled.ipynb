```python
import pandas as pd
import csv
```

```python
df = pd.read_csv('mozilla.csv',
                 index_col='Response ID')
columns = ['FutureFeeling',
              'WiFi Router',
              'Laptop computer',
              'Smart phone',
              'Smart TV',
              'Activity Tracker',
              'Smarthome Hub',
              'Car',
              'Smart Thermostat',
              'Smart Appliance',
              'Smart Door Locks',
              'Smart Lighting' ]

devices = df[columns]
devices = devices.dropna(subset=['FutureFeeling'])
devices.head()
```

```python
feelings = []

for index, row in devices.iterrows():
    if row[0] not in feelings:
        feelings.append(row[0])
        
feelings
```

```python
feelingFrames = []

for feel in feelings:
    feelingFrames.append(devices.loc[df['FutureFeeling'] == feel])
    
allCounts = []
for t in feelingFrames:
    newDf = t.count().values.tolist()
    allCounts.append(newDf)
    
finalCsv = pd.DataFrame(allCounts, columns=columns)
finalCsv
```

```python
# Find the name of the column by index
n = finalCsv.columns[0]

# Drop that column
finalCsv.drop(n, axis = 1, inplace = True)

# Put whatever series you want in its place
finalCsv[n] = feelings

finalCsv
```

```python
finalCsv.to_csv('devices.csv')
```
